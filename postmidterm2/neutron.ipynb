```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("dark_background")
```

```python
qbbc = pd.read_csv('qbbc_100neut_100mev.csv', comment = '#', names = ["X", "Y", "Z", "Energy", "Particle", "Volume", "Process"])
berthp = pd.read_csv('qgspberthp_100neut_100mev.csv', comment = '#', names = ["X", "Y", "Z", "Energy", "Particle", "Volume", "Process"])
bichp = pd.read_csv('qgspbichp_100neut_100mev.csv', comment = '#', names = ["X", "Y", "Z", "Energy", "Particle", "Volume", "Process"])
inclxxhp = pd.read_csv('qgspinclxxhp_100neut_100mev.csv', comment = '#', names = ["X", "Y", "Z", "Energy", "Particle", "Volume", "Process"])
inclxx = pd.read_csv('qgspinclxxx_100neut_100mev.csv', comment = '#', names = ["X", "Y", "Z", "Energy", "Particle", "Volume", "Process"])
```

## Energy

```python
enqbbc = qbbc.groupby(["Volume"]).Energy.sum().reset_index().sort_values(by = ["Energy"], ascending = False)
enberthp = berthp.groupby(["Volume"]).Energy.sum().reset_index().sort_values(by = ["Energy"], ascending = False)
enbichp = bichp.groupby(["Volume"]).Energy.sum().reset_index().sort_values(by = ["Energy"], ascending = False)
eninclxxhp = inclxxhp.groupby(["Volume"]).Energy.sum().reset_index().sort_values(by = ["Energy"], ascending = False)
eninclxx = inclxx.groupby(["Volume"]).Energy.sum().reset_index().sort_values(by = ["Energy"], ascending = False)
```
