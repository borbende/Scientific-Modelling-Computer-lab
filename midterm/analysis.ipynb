```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.genfromtxt("_bis.csv", delimiter = ',', comments = '#')    #Load the dataset.
```

```python
print(np.shape(data))                                                #Checking some properties.
print(data)
```

```python
kep = plt.figure(figsize = (15,10))                                   #Plotting the results.                    

plt.subplot(2,2,1)
plt.hist(data[:,0], bins = 20, color = "gold")
plt.xlabel("Event numbers", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.grid()

plt.subplot(2,2,2)
plt.hist(data[:,1], bins = 100, color = "gold")
plt.xlabel("X positions [mm]", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.grid()

plt.subplot(2,2,3)
plt.hist(data[:,2], bins = 100, color = "gold")
plt.xlabel("Y positions [mm]", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.grid()

plt.subplot(2,2,4)
plt.hist(data[:,3], bins = 100, color = "gold")
plt.xlabel("Z positions [mm]", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.grid()


plt.suptitle("Position of photons upon reaching the sensitive detectors", fontsize = 13)
plt.show()
```

```python
plt.hist(data[:,0], bins = 20, color = "gold")                  #Saving the histogram.
plt.xlabel("Event identification numbers", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.title("Distribution of events", fontsize = 13)
plt.grid()
plt.savefig("histo.png")
```

```python
n, bins, patches = plt.hist(data[:,0], bins = 10, color = "gold")     #Converting data for calculation.
plt.xlabel("Event identification numbers", fontsize = 13)
plt.ylabel("Counts", fontsize = 13)
plt.title("Distribution of events", fontsize = 13)
plt.grid()
```

```python
print(np.mean(n), "+/-", np.std(n))                         #Calculation of mean and standard deviation.
```

```python
plt.plot(data[:,1], data[:,2], linestyle = '', marker = 'o', color = "gold")     #Plotting the phase space.
plt.xlabel("X positions [mm]", fontsize = 13)
plt.ylabel("Y positions [mm]", fontsize = 13)
plt.title("XY phase space", fontsize = 13)
plt.grid()
plt.savefig("xy_phase.png")
```
